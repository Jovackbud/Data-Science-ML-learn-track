```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

Functional method

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y) 
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('1st Plot')
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'r')

plt.subplot(1,2,2)
plt.plot(x,y, 'b')
```

```python
plt.subplot(2,2,1)
plt.plot(x,y, 'r')

plt.subplot(2,2,2)
plt.plot(x,y, 'b')

plt.subplot(2,2,3)
plt.plot(x,y, 'go--')
```

Object Oriented method

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes.plot(x, y)
axes.set_xlabel('X-Axis')
axes.set_ylabel('Y-Axis')
axes.set_title('Title')
plt.show()
```

```python
fig1 = plt.figure()

axes1 = fig1.add_axes([0.1, 0.1, 0.8, 0.8])
axes1.plot(x,y)
axes1.set_title('Larger')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')

axes2 = fig1.add_axes([0.2, 0.5, 0.4, 0.35])
axes2.plot(y,x)
axes2.set_title('Inset')
axes2.set_xlabel('X1')
axes2.set_ylabel('Y1')
plt.show()
```

##### 

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
for current_ax in axes:
    current_ax.plot(x,y)

plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y)
axes[0].set_xlabel('X1')
axes[0].set_ylabel('Y1')
axes[0].set_title('T1')

axes[1].plot(y,x)
axes[1].set_xlabel('X2')
axes[1].set_ylabel('Y2')
axes[1].set_title('T2')
plt.tight_layout()
```

```python
fig,axes = plt.subplots(nrows=1, ncols=4)
axes[0].plot(x,y)
axes[0].set_xlabel('X0')
axes[0].set_ylabel('Y0')
axes[0].set_title('T0')

axes[1].plot(y,x)
axes[1].set_xlabel('X1')
axes[1].set_ylabel('Y1')
axes[1].set_title('T1')

axes[2].plot(x,x**2)
axes[2].set_xlabel('X2')
axes[2].set_ylabel('Y2')
axes[2].set_title('T2')

axes[3].plot(y,y**2)
axes[3].set_xlabel('X3')
axes[3].set_ylabel('Y3')
axes[3].set_title('T3')

plt.tight_layout()
```

```python
fig = plt.figure()
axes_q = fig.add_axes([0, 0, 1, 1])
axes_q.plot(x, x ** 2, label='x squared')
axes_q.plot(x, x ** 3, label='x cubed')
axes_q.legend(loc=0)
```

```python
fig = plt.figure()
axes_q = fig.add_axes([0, 0, 1, 1])
axes_q.plot(x, y, label='x squared', lw=2, ls='--', marker = 's', markeredgecolor = 'green', 
            markerfacecolor = 'purple', markersize = 15, markeredgewidth = '3')
axes_q.set_xlim([0, 2])
axes_q.set_ylim([0, 4])

axes_q.legend(loc=0)
plt.show()
```
