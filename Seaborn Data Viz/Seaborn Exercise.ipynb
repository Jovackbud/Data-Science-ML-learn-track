# Data Visualisation with Seaborn
___


```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# seaborn plot styles
sns.set_style('whitegrid')
```

```python
# seaborn in pre available data sets
titanic = sns.load_dataset('titanic')
```

# EDA

```python
titanic.head()
```

```python
titanic.info()
```

```python
titanic.describe(include='all')
```

### Plots 

```python
# Fare and age
sns.jointplot(x='fare', y='age', data=titanic)
```

```python
#Fare
plt.figure(figsize=(15,14))
sns.displot(titanic['fare'], kde=False, color='r', bins=20)
```

```python
# class and age
sns.boxplot(x='class', y='age', data=titanic, hue='alive')
```

```python
# class and age
sns.swarmplot(x='class', y='age', data=titanic, size=3.5)
```

```python
#Sex
sns.countplot(x=titanic['sex'])
```

```python
# Sex hue='alive'
sns.countplot(data=titanic, x='sex', hue='alive')
```

```python
# correlation and heatmap
titanic_correlation = titanic.corr()
sns.heatmap(titanic_correlation, cmap='coolwarm')
```

```python
g = sns.FacetGrid(data=titanic, col='sex')
g.map(plt.hist, 'age')
```
