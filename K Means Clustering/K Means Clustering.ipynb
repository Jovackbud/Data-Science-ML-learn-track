```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_features=2, n_samples=200, centers=4, cluster_std=1.8, random_state=101)
```

```python
plt.scatter(data[0][:,0], data[0][:,1], c=data[1], cmap='rainbow')
```

```python
data[1]
```

```python
data[0]
```

```python
from sklearn.cluster import KMeans
```

```python
kmean = KMeans(n_clusters=4)
```

```python
kmean.fit(data[0])
```

```python
kmean.cluster_centers_
```

```python
kmean.labels_
```

```python
fig, (axes1, axes2) = plt.subplots(1, 2, sharey=True, figsize = (10,6))
axes1.set_title('K Means')
axes1.scatter(data[0][:,0], data[0][:,1], c = kmean.labels_, cmap='coolwarm')

axes2.set_title('Original')
axes2.scatter(data[0][:,0], data[0][:,1], c = data[1], cmap='coolwarm')
```
